```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
def knn(n):
    # import loan data
    df = pd.read_csv('s3://helen-ml/week-3/iris.csv')

    # split the data into training and test sets
    train, test = train_test_split(df, test_size=0.3)

    # define kNN model
    nbrs = KNeighborsClassifier(n_neighbors=n)

    # "fit" our model
    nbrs.fit(train[['f1','f2','f3','f4']], train['species'])

    # calculate prediction on the test set
    predictions = nbrs.predict(test[['f1','f2','f3','f4']])

    # calculate our accuracy
    return accuracy_score(test['species'], predictions)
```

```python
acc = knn(3)
print('Accuracy: ', acc)
```

```python
plot_x = list(range(2, 20))
plot_y = list()

for x in plot_x:
    acc = knn(x)
    plot_y.append(acc)
    
plt.plot(plot_x, plot_y)
plt.show()
```
